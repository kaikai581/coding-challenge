```python
def jump(nums):
    cur_pos = 0
    nsteps = 0
    while cur_pos < len(nums):
        best_advance = 1
        farthest_reach = -1
        for j in range(1, nums[cur_pos]+1):
            if cur_pos+j >= len(nums):
                best_advance = len(nums)-1-j
                break
            if cur_pos+j+nums[cur_pos+j] > farthest_reach:
                farthest_reach = cur_pos+j+nums[cur_pos+j]
                best_advance = j
        print(farthest_reach, best_advance, cur_pos)
        cur_pos += best_advance
        if cur_pos < len(nums): nsteps += 1
    return nsteps
```

```python
print(jump([2,3,1,1,4]))
print(jump([2,3,0,1,4]))
print(jump([3,2,1]))
```
